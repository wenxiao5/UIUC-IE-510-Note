## (b)

$$(n, d, m) = (100, 30, 5)$$

```python
n=100
d=30
m=5
```

```python
def sigma(x):
    if x<0:
        return exp(x)-1
    else:
        return x
```

```python
Train=Matrix(n,d+1, lambda i,j: normalvariate(0, 1))
```

```python
x_train=Train[:,:-1]
y_train=Train[:,-1]
```

Projection onto a 2-dimensional space: $v=x_1[1,0,0,...,0]^T_{m\times 1}$, $W=x_2[[1,1,...,1]_{1\times 30},0_{1\times 30},0_{1\times 30},0_{1\times 30},0_{1\times 30}]^T_{m\times d}$

$$\sum_{i=1}^{n}\left(y_{i}-v^{T} \sigma\left(W x_{i}\right)\right)^{2}=\sum_{i=1}^{n}\left(y_{i}-x_1\sigma(x_2 sum(x_i))\right)^{2}$$

```python
def f3(x1,x2):
    fv=0
    for i in range(n):
        fv+=(y_train[i][0]-x1*sigma(x2*sum(x_train[i])))**2
    return fv
```

```python
plot3d(f3, (x1,-10,10), (x2,-10,10), aspect_ratio=[1,1,1])
```

```python
contour_plot(f3, (-10,10), (-10,10))
```
